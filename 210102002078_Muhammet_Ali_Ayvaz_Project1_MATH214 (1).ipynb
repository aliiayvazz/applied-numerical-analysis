C) Bisection Method

a,b parameters are (first x value, last x value)

```python
def bisection(a, b, real_val): #that find roots with bisection method a,b = 2, 4
    error = 0.000000000001 #error tolerance
    n = 1 # iteration count
    print("***********THAT'S BISECTION METHOD *************")
    while (1): # always true
        new_root = (a + b) / 2 # that finds new root by calculating (a + b) / 2
        f_a = fx_function(a) # it is f(a)
        f_b = fx_function(b) # it is f(b)
        f_newroot = fx_function(new_root) #it is f(a + b / 2)
        if (f_newroot * f_a < 0): # if f(a) x f(a+b / 2)  < 0  , it is changed b
            b = new_root
        else: #if not changed a
            a = new_root
        if ((b-a) / (2**n) < error): #if the expression < error , then it is founded.
            print("it is founded...\nthe x{} : {}".format(n, new_root))
            print("iteration number : {} (with error : {})".format(n, error))
            my_list.append(["BISECTION", new_root, real_val])
            break
        else:# if not , the loop is continue
            print("{}. iteration \n".format(n))
            print("[  X{} = {}  ]\n".format(n, new_root))
            n += 1 # iteration number changed
            continue
            
bisection(2, 4, 3.6)
```

That function takes one parameter as xn0(according to Newton Raphson methot formula) and starts infinite loop up to real root. For [2,4] Newton Raphson for x1: output:

```python
def newton_method(xn0,real_val): # that find root of f(x) according to newton raphson method
    def formula(xn): ## that calculates x1 with f(x) and derivative of f(x)
        xn1 = xn - fx_function(xn) / d_fx(xn)
        return xn1
        
    error = 0.000000000001 # error tolerance
    n = 1 #iteration count
    print("***********THAT'S NEWTON RAPHSON METHOD*************")
    xn = xn0                          #first values
    xn1 = formula(xn) # calculates x1          
    while (1): #infinite loop
        if (abs(xn1 - xn) < error): # it is calculated absolute(xn1 - xn2)
            print("\nit is founded.... with error tolerance : {}\n".format(error))
            print("iteration number : {}".format(n))
            print("x{} = {}".format(n, xn1))
            my_list.append(["NEWTON RAPHSON", xn1, real_val])
            break  #if yes stops....
        else: # if no, continues
            print("{}. iteration".format(n))
            print("[  x{} = {}   ]".format(n, xn1))
            print("\n")
            n += 1 #iteration number increase + 1
            temp = xn1                        #### swapping old values to new values
            xn = xn1
            xn1 = formula(temp)
            
            
            
newton_method(2, 3.6) #function call to print
```

Secant Methot Functions takes two parameters as x0 and x1 (firstvalues according to formula of secant method) and calculates candidate root unlike Newtob Raphson method. x1 means x(-1) = 1 and x(0) = 2

```python
def secant_method(x1, x0, real_val): #that function takes x0, x1 from user and calculates the root approximately..
    def formula(x0, x1): #like newton raphson method but there are f(x) process after divison instead of F'(x)
        x2 = x1 - (fx_function(x1) * (x1 - x0)) / (fx_function(x1) - fx_function(x0))
        return x2
        
    error = 0.000000000001 # error tolerance
    n = 0 #iteration count
    print("***********THAT'S SECANT METHOD*************")
    x0, x1 = x0, x1 #swapping for other loop.
    x2 = formula(x0, x1) #new root
    while (1): #infinite loop
        if (abs(x2 - x1) < error): #if yes, it stops
            print("\nIt is found with error tolerance: {}\n".format(error))
            print("Iteration number: {}".format(n + 1))
            print("x{} = {}".format(n + 1, x2))
            my_list.append(["SECANT", x2, real_val])
            break
        else: #if not , continues for other candidate values 
            n += 1
            print("{}. iteration".format(n))
            print("[x{} = {}]]".format(n, x2))
            x0, x1 = x1, x2 #swapping for other loop
            x2 = formula(x0, x1) #new candidate root
    
secant_method(2, 3, 3.6) #function call x(-1) = 1, x(0) = 2
```

```python
secant_method(6, 7, 7.4) #function call x(-1) = 6, x(0) = 7
```

```python
def g1(x):
    y = (0.25*x**2 + 6.66) / 2.75
    return y

def g2(x):
    y = (2.75*x - 6.66)**0.5
    return y

#it will be used only one....

def fixed_point_i_method(a, b, function_no, real_val): # a and b mean first x and last x value. function_no means g1 or g2 ???
    print("********THAT'S FIXED POINT METHOD*************************")
    error = 0.000000000001 #error tolerance
    if (function_no == 1): #if function_no 1 , it will be chosen g1 . vice versa
        g = g1
    else:
        g = g2
        
    def condition_test(xi, xi1): # it was written to find any suitable value according to fixed point method formula 
            if (0 < abs((g(xi) - g(xi1)) / (xi - xi1)) < 1 ):  # abs is absolute 
                return xi1
            else:
                return None
    xi = random.uniform(a, b) # it is chosen random any float value between [a, b] by using random library
    xi1 = g(xi) # either g1 or g2.
    while (1): #infinite loop
        if (condition_test(xi, xi1)): # if  0 < .. < 1 , stops if not, continues untill it will be done
            break
        xi = random.uniform(a, b)
        xi1 = g(xi)
    # chosed real value.....
    n = 1 #iteration counter
    print("it is chosed random (x0) : {} for [{}, {}] (because it is suitable according to 0 < ....  < 1 formula) ".format(xi, a, b))
    while (1): #infinite loop
        if (abs(xi1 - xi) < error): #if expression suits < error 
            print("\nIt is found with error tolerance: {}\n".format(error))
            print("x{} = {}".format(n, xi1))
            print("the iteration number : {}".format(n))
            my_list.append(["FIXED POINT", xi1, real_val])
            break #it breaks
        else:
            print("{}. iteration".format(n))
            print("[ x{} = {}]  ]".format(n, xi1))
            xi = xi1
            xi1= g(xi1)
            n +=1 #adds +1 iteration counter
#section d

fixed_point_i_method(2, 4, 1, 3.6) #function call [2, 4] first function
```
